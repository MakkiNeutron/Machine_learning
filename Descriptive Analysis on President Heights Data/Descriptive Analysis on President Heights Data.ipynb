```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/Data science/My projects/Data Analysis and Machine learning on President Heights Data/president_heights.csv')

df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
#Total Number of missing values
df.isnull().sum().sort_values(ascending=False)
```

```python
# check out the rows with missing values
df[df.isnull().any(axis=1)].head()
```

```python
#remove negative values with rows
df=df.dropna()
```

```python
fig_dims = (10, 6)
fig, ax = plt.subplots(figsize=fig_dims)
b =sns.scatterplot(x='height(cm)', y='name',data=df,hue='height(cm)',palette='hls')
b.set_ylabel("US PRESIDENTS",fontsize=14)
b.set_xlabel("HEIGHT IN CENTIMETRE",fontsize=14)
```

```python
#setting up a theme
b= sns
sns.set_theme(style="whitegrid") 
colors=sns.color_palette("hls")

```

```python

print("10th percentile of Height=", df['height(cm)'].quantile(0.1))
print("25th percentile of Height=", df['height(cm)'].quantile(0.25))
print("50th percentile of Height =", np.median(df['height(cm)']))
print("75th percentile of Height=", df['height(cm)'].quantile(0.75))
print("90th percentile of Height=", df['height(cm)'].quantile(0.90))
```
