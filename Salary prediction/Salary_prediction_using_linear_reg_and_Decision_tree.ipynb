```python
x= df[['YearsExperience']]
y= df[['Salary']]
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeRegressor
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
LR = LinearRegression()
DTR = DecisionTreeRegressor()
```

```python
LR.fit(X_train,y_train)
DTR.fit(X_train,y_train)
```

```python
pred_ylg = LR.predict(X_test)
pred_ydtr = DTR.predict(X_test)
```

```python
score_lg=round(LR.score(X_test,y_test)*100,2)
score_dtr=round(DTR.score(X_test,y_test)*100,2)
print('Linear Regression Accuracy :',score_lg,'%')
print('Decision Tree Regressor Accuracy :',score_dtr,'%')
```

```python
plt.scatter(X_train,y_train,color='blue')
plt.plot(X_train,LR.predict(X_train),color='black')
plt.title('Salary vs Years of Experience (Training Data)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary of an employee')
plt.show()

```

```python
plt.scatter(X_test,y_test,color='green')
plt.plot(X_test,LR.predict(X_test),color='black')
plt.title('Salary vs Years of Experience (Test Data)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary of an employee')
plt.show()
```

```python
from sklearn import metrics
print('Mean Absolute Error of the Model:',metrics.mean_absolute_error(y_test,pred_ylg))
print('Mean Squared Error of the Model: ',metrics.mean_squared_error(y_test,pred_ylg ))
print('Root Mean Squared Error of the Model: ',np.sqrt(metrics.mean_absolute_error(y_test,pred_ylg)))
```

```python

print('Mean Absolute Error of the Model:',metrics.mean_absolute_error(y_test,pred_ydtr))
print('Mean Squared Error of the Model: ',metrics.mean_squared_error(y_test,pred_ydtr ))
print('Root Mean Squared Error of the Model: ',np.sqrt(metrics.mean_absolute_error(y_test,pred_ydtr)))
```
